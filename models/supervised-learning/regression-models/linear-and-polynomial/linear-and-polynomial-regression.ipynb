```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("Position_Salaries.csv")
```

```python
df = df.iloc[:, 1:3]
print(df.head())
print(df.info())
print(df.columns)
```

```python
X = df[['Level']].values # Level column ->2d array
y = df['Salary'].values  # Salary columns
plt.scatter(X, y, color='red')
plt.plot(X, y, color='red')
plt.show()
```

```python
# linear regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# polynomial reg.
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, poly_reg.predict(X_poly),color='maroon')
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Linear regression prediction
print("Linear Regression Prediction for Level 6.5:\t",
      lin_reg.predict([[6.5]]) [0])
```

```python
# Polynomial regression prediction
print("Polynomial Regression Prediction for Level 6.5:\t",
      poly_reg.predict(poly.transform([[6.5]])) [0])
```
